```python
# 기준별 평균 월간 판매량을 계산해 새로운 컬럼으로 추가하는 함수 정의
def mean_cnt_month(data, features):
    
    # 새로 생성될 피처명 생성
    new_feature_name = '/'.join(features[1:])+'_mean_cnt'
    
    # 기준에 따라 월간 평균 판매량 구하기
    pivot = data.pivot_table(index=features, values='cnt_month', aggfunc='mean')
    pivot = pivot.reset_index().rename(columns={'cnt_month': new_feature_name})
    
    # data와 pivot 병합 
    data = data.merge(pivot, on=features, how='left')
    
    del pivot

    return data

train_5 = mean_cnt_month(train_4,['month_id','item_id'])

train_5.head()
```

```python
# 각 매장/상품 별로 임의로 정한 개월 수 이전의 값으로 컬럼을 추가하는 함수 정의
# num_lags=2라면 1개월전의 값 컬럼과 2개월전의 값 컬럼이 생성됨
# rolling()과 같은 과정이지만 현 데이터에 rolling을 적용하기 복잡하기 때문에 함수로 정의해서 사용

def make_lags(data, lag_feature, num_lags):
    temp = data[['month_id','shop_id','item_id',lag_feature]].copy() # 원본은 건드리지 않도록 copy()

    # lags 피쳐 생성 및 기존 데이터와 병합
    for i in range(1, num_lags+1):
        column_name = f'{lag_feature}_lag_{str(i)}' # 새로 생성될 컬럼명 지정 
        temp.rename(columns={lag_feature:column_name}, inplace=True) # temp 컬럼명 새 이름으로 재지정

        temp['month_id'] += i # month_id에 해당 lag 숫자(몇 개월 전인지)를 더하기 > 해당 숫자만큼 월이 밀린 데이터가 됨 
        
        data = data.merge(temp, on=['month_id','shop_id','item_id'], how='left')
        
        # df[column_name] = df[column_name].fillna(0) # 판매이력이 없는 경우(신상품) NaN이 존재할 수 있으므로 결측치는 0으로 대체
   
    del temp

    return data
```
