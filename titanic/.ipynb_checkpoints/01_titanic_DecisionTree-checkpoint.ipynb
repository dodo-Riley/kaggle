```python
# fare 컬럼의 이상치 데이터 확인

## 이상치 값을 찾아주는 함수 정의
def outliers(data,col):
    Q1,Q3 = np.percentile(data[col], [25,75])
    IQR = Q3-Q1
    
    bound_upper = Q3+1.5*IQR
    bound_lower = Q1-1.5*IQR
    
    outliers_upper = data[data[col]>bound_upper]
    outliers_lower = data[data[col]<bound_lower]
    
    df_outliers = pd.concat([outliers_upper,outliers_lower])
    df_outliers = df_outliers.sort_values(by=col, ascending=False)
    
    return df_outliers

## 이상치만 모은 데이터프레임 생성
fare_outliers = outliers(df_except,'fare')
fare_outliers
```
